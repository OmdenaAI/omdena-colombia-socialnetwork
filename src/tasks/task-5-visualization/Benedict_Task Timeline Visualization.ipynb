```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import numpy as np
import datetime
from jupyterthemes import jtplot

jtplot.style('gruvboxd')
%matplotlib inline
```

```python
task_zero = ['Challenge onboarding', 'Domain knowledge', ]
task_zero_dates = ['2021-08-12', '2021-08-23', ]
```

```python
task_one = ['Data scraping & consolidation (Twitter)', 'Data scraping (Reddit)',
            'Data rescraping & consolidation (Twitter)', 'Data consolidation (Reddit)',
            'Data rescraping & storage expansion']
task_one_dates = ['2021-08-30', '2021-08-31', '2021-09-05', '2021-09-05', '2021-09-15']
```

```python
task_two = ['Preprocessing (English)', 'Preprocessing (Spanish)',
            'Data aggregation (English & Spanish)', ]
task_two_dates = ['2021-09-05', '2021-09-10', '2021-09-14', ]
```

```python
task_three = ['Language detection', 'Topic modelling (EDA)', 'Topic modelling & interpretation',
              'COREX-BERT-NMF algorithm trials', 'Language translation']
task_three_dates = ['2021-09-10', '2021-09-15', '2021-09-20', '2021-09-25', '2021-09-29']
```

```python
task_four = ['Sentiment generation', 'Preliminary modelling']
task_four_dates = ['2021-09-12', '2021-09-15', ]
```

```python
task_five = ['Streamlit dashboard', 'Wordcloud animation',
             'Barplot animation', 'Sentiment barplot', ]
task_five_dates = ['2021-09-23', '2021-09-27', '2021-10-03', '2021-10-05',]
```

```python
others = ['DagsHub walkthrough', 'DagsHub workshop', ]
others_dates = ['2021-09-07', '2021-09-30', ]
```

```python
all_tasks = task_zero + task_one + task_two + task_three + task_four + task_five + others
all_dates = task_zero_dates + task_one_dates + task_two_dates + task_three_dates + task_four_dates + task_five_dates + others_dates
all_dates = [datetime.datetime.strptime(date, '%Y-%m-%d') for date in all_dates]
```

```python
s = '-' * 100
```

```python
levels = np.tile([-7, 7, -5, 5, -4, 4, -3, 3, -2, -1, 1], int(np.ceil(len(all_dates)/6)))[:len(all_dates)]

fig, ax = plt.subplots(figsize = (100, 50), constrained_layout = True)
ax.set_title(f'Timeline for Omdena Colombia National Strike Challenge\n{s}', fontsize = 150, pad = 80,
             fontfamily = 'bell mt')

ax.vlines(all_dates, 0, levels, linewidth = 5, color = 'green')
ax.plot(all_dates, np.zeros_like(all_dates), '-D', color = '#ffffff', markerfacecolor = 'r',
        markeredgecolor = 'green', markeredgewidth = 7, markersize = 100)

for d, l, r in zip(all_dates, levels, all_tasks):
    ax.annotate(r, xy = (d, l), xytext = (-2, np.sign(l)*8), fontfamily = 'bell mt',
                textcoords = 'offset points', fontsize = 110,
                horizontalalignment = 'left' if (r.__contains__('Challenge') |
                                                 r.__contains__('Sentiment generation') |
                                                 r.__contains__('Streamlit') |
                                                 r.__contains__('detection')) else 'right',
                verticalalignment = 'bottom' if l > 0 else 'top'
               )
    ax.plot(d, l, 'D',  markeredgecolor = 'green', markeredgewidth = 5,
            markerfacecolor = 'maroon', markersize = 100)
    
ax.xaxis.set_major_locator(mdates.DayLocator(interval = 4))
ax.xaxis.set_major_formatter(mdates.AutoDateFormatter('%b %Y'))

plt.setp(ax.get_xticklabels(which = 'both'), fontsize = 60, rotation = 0, ha = 'right')

ax.yaxis.set_visible(False)
ax.spines[['left', 'right', 'top']].set_visible(False)
ax.margins(y = 0.1, tight = True)
plt.savefig('Omdena Colombia Timeliner.pdf', dpi = 200)

plt.show(); plt.close('all')
```
